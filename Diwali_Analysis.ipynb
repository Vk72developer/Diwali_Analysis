# **Diwali_Data_Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data Collection & Data Cleaning

```python
data_sets = pd.read_csv('/content/Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
data_sets
```

# Exploratory Data Analysis

```python
data_sets.columns
```

Gender

```python
# Without Labels
ax = sns.countplot( x = 'Gender', data = data_sets)
```

```python
# With labels
ax = sns.countplot( x = 'Gender', data = data_sets)

for bars in ax.containers:
   ax.bar_label(bars)
```

```python
data_sets.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = data_sets.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
```

**From above the graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men**

Age

```python
data_sets.columns
```

```python
# count of age - group
ax = sns.countplot(data = data_sets, x = 'Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age group

sales_age = data_sets.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)

sns.barplot( x = 'Age Group', y = 'Amount', data = sales_age)
```

**From above the graphs we can see the most of the buyers are of age group between 26 - 35 yrs female**

State

```python
# total no. of orders from top 10 states

sales_state = data_sets.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15, 4)})
sns.barplot( data = sales_state, x = 'State', y= 'Orders')
```

```python
# total amount/ sales from top 10 states

sales_state = data_sets.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15, 4)})
sns.barplot( data = sales_state, x = 'State', y= 'Amount')
```

**From above the graphs we can see that unexpectedly most of the orders are U.P, Maharastra and Karnatka respectively**

Marital Status

```python
# Count

ax = sns.countplot(data = data_sets, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data_sets.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6, 5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue='Gender')
```

**From above the graphs we can see the buyers are married (women) and they have high purchasing power**

Occupation

```python
sns.set(rc={'figure.figsize': (23, 5)})

ax = sns.countplot(data = data_sets, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data_sets.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (25, 5)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

From above the graphs we can see the most of the buyers are working in IT, Healthcare and Aviation sector

Product Category

```python
sns.set(rc={'figure.figsize': (30, 10)})

ax = sns.countplot(data = data_sets, x = 'Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = data_sets.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize': (25, 5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')
```

From above the graphs we can see the most of the sold products are from food, clothing and electronics category

```python
sales_state = data_sets.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```

```python
# top 10 most products

fig1, ax1 = plt.subplots(figsize=(12, 7))
data_sets.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

Conclusion

Married women age group 26 - 35 from UP, Maharastra and karnatka working in IT , HealthCare and Aviation are more likely to buy products from food , clothing and Electronics category
